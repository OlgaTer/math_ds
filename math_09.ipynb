```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистрограммы средних этих выборок.

```python
# создадим случайную величину из распределения "хрустальный шар"
# (beta > 0 - определяет точку, в которой pdf меняется от степенного распределения к гауссовому,
# m > 1 - определяет силу степенного "хвоста")
beta, m = 3, 5
rv = sts.crystalball(beta, m)
sample = rv.rvs(size=1000)
```

```python
# построим график по исходным параметрам
x = np.linspace(0, 2, 100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
# Параметры запуска: размеры выборок и количество повторов генерации
n1, n2, n3, n4 = 2, 5, 15, 35

samples_count = 1000
# Массивы для сохранения средних по выборкам
a1 = np.array([])
a2 = np.array([])
a3 = np.array([])
a4 = np.array([])

for number in range(samples_count):
    # На каждом цикле генерируем выборки разных размеров
    sample1 = rv.rvs(size=n1)
    sample2 = rv.rvs(size=n2)
    sample3 = rv.rvs(size=n3)
    sample4 = rv.rvs(size=n4)

    # Считаем среднее по каждой выборке и добавляем в массив средних
    a1 = np.append(a1, sample1.mean())
    a2 = np.append(a2, sample2.mean())
    a3 = np.append(a3, sample3.mean())
    a4 = np.append(a4, sample4.mean())
```

```python
# Строим графики
# Подписи для графиков
plt.figure(figsize=(12,8))

plt.subplot(2, 2, 1)
label_1 = 'sample of ' + str(n1) + '\n histogram'
plt.hist(a1, density=True, label=label_1)
plt.legend(loc='center left')
plt.ylabel('samples means')
plt.xlabel('$x$')

plt.subplot(2, 2, 2)
label_2 = 'sample of ' + str(n2) + '\n histogram'
plt.hist(a2, density=True, label=label_2)
plt.legend(loc='center left')
plt.ylabel('samples means')
plt.xlabel('$x$')

plt.subplot(2, 2, 3)
label_3 = 'sample of ' + str(n3) + '\n histogram'
plt.hist(a3, density=True, label=label_3)
plt.legend(loc='center left')
plt.ylabel('samples means')
plt.xlabel('$x$')

plt.subplot(2, 2, 4)
label_4 = 'sample of ' + str(n4) + '\n histogram'
plt.hist(a4, density=True, label=label_4)
plt.legend(loc='center left')
plt.ylabel('samples means')
plt.xlabel('$x$')
plt.show()
```

При увеличении **n** график средних значений выборок растет в высоту, одновременно сужаясь по оси Х. Он постепенно приближается к графику нормального распределения случайных величин, а это подтверждает, что центральная предельная теорема работает.
