#### Уровень 0:

* Создайте матрицу 3 на 3, заполненную числами от 1 до 9
* Прибавьте к ней матрицу с единицами на диагонале
* Умножьте полученную матрицу на вектор (1, 2, 3)
* Посчитайте обратную матрицу к полученной матрице

```python
import numpy as np
```

```python
# создадим требуемую матрицу: преобразуем массив чисел от 1 до 9 
# в матрицу размером ЗхЗ

a = np.arange(1,10).reshape(3,3)
a
```

```python
# сложим исходную матрицу с единичной диагональной (той же размерности)

b = a + np.eye(3)
b
```

```python
# умножим новую матрицу на вектор (1,2,3)

c = b @ np.array([1,2,3])
c
```

```python
# найдем матрицу, обратную матрице b

d = np.linalg.inv(b)
d
```

```python
# проверим, что матрица d обратна матрице b
# (их скалярное произведение даст единичную диагональную матрицу)

np.dot(b,d)
```

#### Уровень 1:

Дан объект в 2d пространстве:

```python
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.xlim([-200, 200])
plt.ylim([-200, 200])

plt.grid()
```

При помощи матричных операций выполните:

* Уменьшить объект в два раза
* Повернуть на 130 градусов
* Отразить объект относительно прямой y=x
* Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1.2)

уменьшим фигуру в 2 раза

```python
# сначала создадим матрицу для изменения масштаба

matrix_1 = np.array([[0.5,0,0],
                    [0,0.5,0],
                    [0,0,1]])
matrix_1
```

```python
# умножим исходную матрицу A на матрицу преобразования

B = np.dot(A,matrix_1)
B
```

```python
# изобразим графически новый объект

x1 = B[:,0]
y1 = B[:,1]

plt.plot(x1, y1)
plt.xlim([-150, 150])
plt.ylim([-150, 150])

plt.grid()
```

повернем объект на 130 градусов (по часовой стрелке)

```python
# сначала создадим матрицу поворота

# т.к. угол в 1 градус - это число Пи, деленное на 180,
# то 130 градусов, соответственно, - это (13 х Пи) / 18

angle = (13 * np.pi) / 18

matrix_2 = np.array([[np.cos(angle), -np.sin(angle), 0],
                    [np.sin(angle), np.cos(angle), 0],
                    [0, 0, 1]])
matrix_2
```

```python
# умножим исходную матрицу A на матрицу поворота

C = np.dot(A,matrix_2)
C
```

```python
# изобразим графически новый объект

x2 = C[:,0]
y2 = C[:,1]

plt.plot(x2, y2)
plt.xlim([-200, 200])
plt.ylim([-200, 200])

plt.grid()
```

отразим объект относительно прямой y=x (т.е. точка с координатами X,Y преобразуется в точку Y,X)

```python
# создадим новую матрицу преобразований

matrix_3 = np.array([[0,1,0],
                    [1,0,0],
                    [0,0,1]])
matrix_3
```

```python
# умножим исходную матрицу A на матрицу преобразования

D = np.dot(A,matrix_3)
D
```

```python
# изобразим графически новый объект

x3 = D[:,0]
y3 = D[:,1]

plt.plot(x3, y3)
plt.xlim([-200, 200])
plt.ylim([-200, 200])

plt.grid()
```

выполним смещение объекта на (200, 300) с масштабированием на (0.5, 1.2) (применим 2 матрицы трансформаций, т.е. композицию линейных отображений)

```python
# создадим первую матрицу - матрицу смещения

matrix_4 = np.array([[1,0,0],
                    [0,1,0],
                    [200,300,1]])
matrix_4
```

```python
# и вторую - матрицу масштабирования

matrix_5 = np.array([[0.5,0,0],
                    [0,1.2,0],
                    [0,0,1]])
matrix_5
```

```python
# применим к исходной матрице A композицию линейных отображений

E = (np.dot(A,matrix_4)) @ matrix_5
E
```

```python
# изобразим графически полученный объект

x4 = E[:,0]
y4 = E[:,1]

plt.plot(x4, y4)
plt.xlim([0, 300])
plt.ylim([0, 700])

plt.grid()
```

решим одну из задач, например, поворота на 130 градусов, другим способом

```python
# создадим матрицу поворота, размерностью 2х2 (величина угла задана выше)

matrix_0 = np.array([[np.cos(angle), -np.sin(angle)],
                    [np.sin(angle), np.cos(angle)]])
matrix_0
```

```python
# умножим каждый вектор матрицы А на матрицу поворота

# сначала создадим новую матрицу cо значениями первого вектора
C_0 = np.array([np.array([x[0],y[0]]) @ matrix_0])

# добавим в цикле к матрице С_0 новые векторы
for i in range(len(A)-1):
    vector = np.array([x[i+1],y[i+1]]) @ matrix_0
    C_0 = np.vstack((C_0,vector))

# посмотрим на итог
C_0
```

```python
# проверим результат графически (будет тем же, что и при перемножении матриц выше)

x_0 = C_0[:,0]
y_0 = C_0[:,1]

plt.plot(x_0,y_0)
plt.xlim([-200, 200])
plt.ylim([-200, 200])

plt.grid()
```
