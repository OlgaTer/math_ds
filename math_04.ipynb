### Уровень 0:

Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

Первая производная функции:

$5x^4 + 4*cos(2x)*2 - sin(3x + 3)*3 = 5x^4 + 8cos(2x) - 3sin(3x + 3)$

в точке х = 1 будет равна:

$5 + 8cos(2) - 3sin(6) = 5 + 8*(-0.41614683654) - 3*(-0.27941549819) = 2.50907180222$

Вторая производная - это производная от первой производной, поэтому:

$20x^3 - 8 * sin(2x) * 2 - 3 * cos(3x + 3) * 3 = 20x^3 - 16sin(2x) - 9cos(3x + 3)$

в точке х = 1 будет равна:

$20 - 16sin(2) - 9cos(6) = 20 - 16 * 0.90929742682 - 9 * 0.96017028665 = -3.19029140906$

Проверим эти вычисления.

```python
# используем SymPy (библиотеку Python) для выполнения символьных вычислений

from sympy import *
```

```python
# зададим символьные переменные

x, y = symbols('x y')
```

```python
# запишем исходную функцию

f = x**5 + 4*sin(2*x) + cos(3*x+3)
```

```python
# вычислим первую производную

fd = diff(f, x)
fd
```

```python
# вычислим вторую производную

sd = diff(fd, x)
sd
```

```python
# определим значения производных в требуемой точке:
# используем функцию evalf с подстановкой 1 вместо Х

# значение первой производной
print(fd.evalf(subs={x: 1}))
```

```python
# значение второй производной
print(sd.evalf(subs={x: 1}))
```

### Уровень 1:

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

на отрезке $[-5; 5]$.

```python
# зададим исходную функцию

f_1 = sin(2*x+1)**5
```

```python
# найдем производные:

# первую

fd_1 = diff(f_1, x)
fd_1
```

```python
# и вторую

sd_1 = diff(fd_1, x)
sd_1
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# зададим требуемый отрезок (значения по оси Х)

axis_x = np.linspace(-5, 5, endpoint=True)
```

```python
# рассчитаем значения производных в каждой заданной точке

t_1 = []
t_2 = []

for i in range(len(axis_x)):
    r_1 = fd_1.evalf(subs={x: i})
    t_1.append(r_1)

    r_2 = sd_1.evalf(subs={x: i})
    t_2.append(r_2)
```

```python
# построим график

plt.figure(figsize=(8,6))

plt.plot(axis_x, t_1, c='red', label='first derivative')
plt.plot(axis_x, t_2, c='blue', label='second derivative')

plt.legend()
plt.grid(axis='x')
```
